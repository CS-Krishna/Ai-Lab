```python
import numpy as np
import pandas as pd
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print("Dot product:", np.dot(v1, v2))
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("Matrix multiply:\n", A @ B)
```

```python
print("Transpose:\n", A.T)
```

```python
print("Inverse:\n", np.linalg.inv(A))
```

```python
print("Determinant:", np.linalg.det(A))
```

```python
coeff = np.array([[2,1],[1,3]])
rhs   = np.array([5,10])
sol   = np.linalg.solve(coeff, rhs)
print("Solution x,y:", sol)
```

```python
cols = ['sepal_len','sepal_wid','petal_len','petal_wid','class']
df   = pd.read_csv('iris.csv', header=None, names=cols)
nums = df.drop(columns='class')
```

```python
print("Mean:\n",   nums.mean())
```

```python
print("Median:\n", nums.median())
```

```python
print("Standard Deviation:\n",    nums.std())
```

```python
print("Variance:\n",    nums.var())
```

```python
print("Correlation:\n",   nums.corr())
```
