```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ['sepal_len','sepal_wid','petal_len','petal_wid','class']
df = pd.read_csv('iris.csv', header=None, names=cols)
```

```python
means = np.mean(df.iloc[:, :-1], axis=0)
print("Column Means:\n", means)
```

### Plot

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(df['sepal_len'], bins=20, color='steelblue')
plt.title("Sepal Length Histogram")
```

```python
plt.subplot(1,2,2)
for cls in df['class'].unique():
    subset = df[df['class']==cls]
    plt.scatter(subset['sepal_len'], subset['petal_len'], label=cls, s=20)
plt.legend(); plt.title("Sepal vs Petal Length")
plt.tight_layout(); plt.show()
```
