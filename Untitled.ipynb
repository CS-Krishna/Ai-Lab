```python
import pandas as pd
import numpy as np

filepath = r"C:\Users\Krishna\Downloads\archive (3)\cic.csv"

# --- Read in chunks, grab 50k BENIGN and 50k Attack rows ---
benign_rows  = []
attack_rows  = []
benign_limit = 50_000
attack_limit = 50_000

for chunk in pd.read_csv(filepath, chunksize=50_000):
    chunk.columns = chunk.columns.str.strip()
    chunk = chunk.replace([np.inf, -np.inf], np.nan).dropna()
    
    benign = chunk[chunk['Label'].str.strip() == 'BENIGN']
    attack = chunk[chunk['Label'].str.strip() != 'BENIGN']
    
    if len(benign_rows) < benign_limit:
        benign_rows.append(benign)
    if len(attack_rows) < attack_limit:
        attack_rows.append(attack)
    
    # Stop once we have enough of both
    benign_count = sum(len(b) for b in benign_rows)
    attack_count = sum(len(a) for a in attack_rows)
    print(f"Benign: {benign_count:,} | Attack: {attack_count:,}", end='\r')
    
    if benign_count >= benign_limit and attack_count >= attack_limit:
        break
```
